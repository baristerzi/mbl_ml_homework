```python
def puan_ortala(x):
    response=requests.get('https://www.yemeksepeti.com'+x)
    response_bs=bs4.BeautifulSoup(response.text, "html.parser")
    points=response_bs.findAll("span",attrs={'class':"point point10 ys-invert"})
    pointlist=[]
    for i in points:
        i1=i.string
        i2=i1.replace(",",".")
        pointlist.append(i2)
    totalpoint=0
    for i in pointlist:
        totalpoint+=float(i)
    lenpointlist=len(pointlist)
    try:
        avpoint=totalpoint/lenpointlist
    except ZeroDivisionError:
        avpoint=0
        
    return avpoint
```

```python
def min_price(x):
    response=requests.get('https://www.yemeksepeti.com'+x)
    response_bs=bs4.BeautifulSoup(response.text, "html.parser")
    price=response_bs.findAll("div",attrs={'class':"column"})
    pricelist=[]
    cnt=0
    for i in price:
        i1=i.text.strip()
        i2=i1.strip("TL")
        i3=i2.replace(",",".")
        if cnt%3==0:
            pricelist.append(i3)
        else:
            pass
        cnt+=1
    totalprice=0
    for i in pricelist:
        totalprice+=float(i)
    lenpricelist=len(pricelist)
    try:
        avprice=totalprice/lenpricelist
    except ZeroDivisionError:
        avprice=0
        
    return avprice
```

```python
dataset=[]
for i,j in zip(campuses_name,campuses):
    dataset.append(i)
    avp=puan_ortala(j)
    dataset.append(avp)
    avpr=min_price(j)
    dataset.append(avpr)
```

```python
dataset
```

```python
f=open('campus_point.csv','a')
p=0
l=len(dataset)-1
f.write("Campus Name,Average Restaurant Point,Average Minimum Price\n")
while p<l:
    q=0
    while q<2:
        f.write(str(dataset[p]) + ", ")
        p=p+1
        q=q+1
    f.write(str(dataset[p]) +"\n ")
    p=p+1
f.close()
```
